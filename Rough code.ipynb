```python
import pandas as pd
from datetime import datetime

df_read = pd.read_csv('Customer Dataset - Sheet1.csv')
df_read[['TXN','CONST_NUM','RRN','ACC_NUM','BANK','CUST_NAM','TRANS_TYPE']] = df_read['NARRATION'].str.split('/',expand=True)

# Number records
len(df_read)
# mydoc = my_collection.find().count()
# Unique banks
df_read['BANK'].unique()
print('Unique banks:{}'.format(df_read['BANK'].unique()))
# Value between 
df_read['TXN DATE'] = pd.to_datetime(df_read['TXN DATE'])  

end_date = datetime.strptime('2022-07-31','%Y-%m-%d')
start_date = datetime.strptime('2022-07-21','%Y-%m-%d')
mask = df_read[(df_read['TXN DATE'] > start_date) & (df_read['TXN DATE'] <= end_date)]['TXN']
print(len(mask))

# Based on type total sum amount
print(df_read.groupby('TRANS_TYPE').sum())

for names in df_read['CUST_NAM'].unique():
    output = ''.join(x for x in names.title() if x.isalnum())
    print(output[0].lower() + output[1:])

    print(names.title())
```
